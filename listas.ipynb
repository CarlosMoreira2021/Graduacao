```python
mochila = ['Machado','Camisa','Bacon','Abacate']
print('Lista',mochila)
```

trocando item da lista

```python
mochila[2] = 'Laranja'
#vai trocar o bacon pela laranja
print('Lista',mochila)
```

Manupulando listas

```python
#acrescentou os ovos na lista
mochila.append('Ovos')
print('Lista',mochila)
```

```python
#inserindo o item onde eu quero na lista e desloca o indice que estiver no local para direita 
mochila.insert(1,'Canivete')
print('Lista',mochila)
```

```python
#deleta o indice da lista 

del mochila[0]
print('Lista',mochila)

#deleta mas especificando o indice 

mochila.remove('Ovos')
print('Lista',mochila)
```

copia de listas

```python
# mesma referencia

x = [1,10,4,5,6]
y = x

print(x)
print(y)
```

```python
#o y e apenas o endereco que aponta para variavel x, sendo assim, se alterarmos uma vamos alterar x e y 
y[0]=2 

print(x)
print(y)
```

```python
#copia na memoria tornando a variavel y idependente 

x = [5,7,9,11]
y = x[:]
print(x)
print(y)

# agora vamos alterar y para fazer os testes

y[0]=2
print(x)
print(y)
```
