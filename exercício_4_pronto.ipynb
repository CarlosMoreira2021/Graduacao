```python
peca = {}
pecas = []
codigo = -1


def cadastrarPeca():
    global pecas
    global peca
    global codigo

    for i in range(1):

        peca['codigo'] = codigo = codigo + 1
        peca['peça'] = input('Por favor entre com o nome da peça : ')
        peca['fabricante'] = input('Por favor entra com o Fabricante da peça: ')
        peca['valor'] = input('Por favor entre com o VALOR(R$) da peça: ')
        pecas.append(peca.copy())


def consultarPeca():
    global pecas
    global peca
    global codigo
    while True:

        print('1- Consultar todas as pecas')
        print('2- Consultar peças por codigo')
        print('3- Consultar peças por fabricante')
        print('4- sair')

        rConsulta = int(input())

        if rConsulta != 1 and rConsulta != 2 and rConsulta != 3 and rConsulta != 4:
            print('Opcao invalida')
            continue

        elif rConsulta == 4:
            print('Programa finalizado')
            break


        elif rConsulta == 1: # consultar todos os produtos
            for e in pecas:
                print('-' * 20)
                for i, j in e.items():
                    print(i, j)
            print('-' * 20)




        elif rConsulta == 2: # consultar produto por codigo
            r = int(input('digite codigo do produtos:'))
            res = pecas[r]
            print('-' * 20)
            for k, v in res.items():
                print(f'{k}: {v}')
            print('-' * 20)


        elif rConsulta == 3:# consultar produto pro fabricante
            r = str(input('Digite o fabricante:3'))
            for peca in pecas:
                if (peca['fabricante'] == r):
                    for k, v in peca.items():
                        print(f'{k}: {v}')
                print('-' * 20)


        elif rConsulta == 4: #retorna para o menu principal
            break


def removerPeca():
    r = int(input('Digite o codigo do produto para remover a peça do sistema:'))
    for peca in pecas:
        if (peca['codigo'] == r):
            pecas.remove(peca)


while True:
    print('Bem vindo ao Controle de Estoque da Bicicletaria do Carlos Eduardo Moreira de Oliveira')
    print('RU:3615321')
    print('1-Cadastrar Peça')
    print('2-Consultar Peça')
    print('3-Remover Peças')
    print('4-Sair')

    resposta = int(input())

    if resposta != 1 and resposta != 2 and resposta != 3 and resposta != 4:
        print('Opcao invalida')
        continue


    elif resposta == 1:
        # chama funcao cadastrar peca
        print("Voce selecionou a opcao de Cadastrar Peça")
        cadastrarPeca()
        continue

    elif resposta == 2:
        # chama funcao consultar peca
        print("Voce selecionou a opcao de Consultar Peça")
        consultarPeca()
        continue

    elif resposta == 3:
        # chama funcao remover peca
        print("Voce selecionou a opcao de Remover Peça")
        removerPeca()
        continue

    elif resposta == 4:
        print('Programa finalizado')
        break

print('Fim do programa')
```
