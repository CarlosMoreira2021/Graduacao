Estrutura de repeticao



*   Estrutura no programa em que todas as instrucoes contidas nela se repetem de maneira indefinida, ate que uma condicao seja satisfeita

*  Sinonimos: Estrutura iterativa, laco de repeticao ou loop de repeticao 





ESTRUTURA DE REPETICAO WHILE (ENQUANTO)



*   Repete um bloco de instrucoes enquanto determinada condicao se mativer verdadeira
*   Caso contrario, ocorre o desvio para a primeira linha de codigo apos este bloco de repeticao

Acumuladoras

```python
inicial = int(input('Qual valor deseja iniciar a contagem? '))
final = int(input('Qual valor deseja encerrar a contagem? '))

x = inicial
while(x <= final):
  if (x % 2 == 0):
    print(x)
  x= x + 1
```

Escreva um algoritmo que calcule a sua media de notas em determinada disciplina.
Vamos assumir que a media final  dada pela media aritmetica de cinco notas digitadas

```python
soma = 0 
cont = 1
while (cont <= 5):
  x = float(input('Digite a {}  nota: '.format(cont)))
  soma = soma + x
  cont = cont + 1
media = soma / 5
print('Media final: {}'.format(media))
```

Crie um algoritmo que receba um valr do tipo inteiro via teclado no entanto, o programa s deve aceitar, obrigatoriamente, valores inteiros e positivos.
Qualquer valor negativo, ou igual a zero, deve ser rejeitado pelo programa e um novo valor deve ser solicitado.

```python
val = int(input('Digite um valor maior que 0: '))
while (val <= 0):
    val = int(input('Digite um valor maior do que 0: '))
print('Voce digitou {}. Encerrando o programa...'.format(val))
```

Escreva um algoritmo que fique recebendo frases ou palavras digitadas.
Encerre o algoritmo quando a palavra sair for digitada

```python
print('Digite uma mensagem que irei repetir pra voce! ')
print('Para encerrar escreva "sair"')

while True:
  texto = input('')
  print(texto)
  if texto == 'sair' or texto == 'SAIR':
    break
print('Encerrando o programa')
```

Escreva um agoritmo em python que calcule a tabuada de todos os numeros de 1 ate 10,e,para cada numero vamos calcular a tabuada multiplicando- o pelo intervalo de 1 at 10.

```python
#while
tabuada = int(input('Digite um numero de 1 ate 10:'))

while tabuada <= 10:
  print('TABUADA DO {}:'.format(tabuada))
  i = 1
  while i <= 10:
    print('{} x {} = {}'.format(tabuada,i,tabuada * i))
    i += 1
  tabuada += 1
```

```python
#for

for tabuada in range(1,11,1):
  print('TABUADA DO {}'.format(tabuada))
  for i in range(1,11,1):
    print('{} x {} = {}'.format(tabuada,i, tabuada * i))
```
