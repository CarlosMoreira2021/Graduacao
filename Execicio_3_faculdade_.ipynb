```python
print('Bem vindo a Companhia de logistica Carlos Eduardo Moreira de Oliveira ')
print('RU - 3615321')

#valordimensao

##tratamento de erro
while True:
    try:
        a = float(input('Digite a altura do objeto (em cm):'))
        l = float(input('Digite a largura do objeto (em cm):'))
        c = float(input('Digite a comprimento do objeto (em cm):'))
        s = a * l * c
        break
    except ValueError:
        print('Voce digitou alguma dimensao do objeto com valor nao numero')
        continue

##dimensao objeto
def dimensaoObj():
    global s
    global vd
    if s < 1000:
        vd = 10
        print(f'O volume do objeto é em cm3: {s}')
    elif s <= 1000 or s < 10000:
        vd = 20
        print(f'O volume do objeto é em cm3: {s}')
    elif s <= 10000 or s < 30000:
        vd = 30
        print(f'O volume do objeto é em cm3: {s}')
    elif s <= 30000 or s < 100000:
        vd = 50
        print(f'O volume do objeto é em cm3: {s}')
    elif s >= 100000:
        print(f'O volume do objeto é em cm3: {s}')
        print('Não aceitamos objetos com dimensoes tao grandes')
        print('Entre com as dimensoes novamente')
        
vd = 0
dimensaoObj()


#tratamento peso

while True:
    try:
        p = float(input('Digite o peso do objeto (em kg):'))
        break
    except ValueError:
        print('Voce digitou peso do objeto com valor nao numerico')
        continue


def pesoObjeto():
    global p
    global peso
    if p < 0.1:
        peso = 1
    elif p <= 0.1 or p < 1:
        peso = 1.5
    elif p <= 1 or p < 10:
        peso = 2
    elif p <= 10 or p < 30:
        peso = 3
    elif p >= 30:
        print('Nao aceitamos objetos tao pesados')

peso = 0
pesoObjeto()





def rotaObjeto():

  while True:
      print('BR - De Brasilia até Rio de Janeiro ')
      print('BS - De Brasilia até São Paulo')
      print('RB - De Rio de Janeiro até São Paulo')
      print('RS - De Rio de janeiro até São Paulo')
      print('SR - De São Paulo até Rio de Janeiro-')
      print('SB - De São Paulo até Brasilia')
      rt = str(input("Selecione uma rota:").upper())
      global rota
      if rt != 'BR' and rt != 'BS' and rt != 'RB' and rt != 'RS' and rt != 'SR' and rt != 'SB':
        print('voce digitou uma rota que nao existe')
        continue
      elif rt == 'BR':
        rota += 1.5
        print('Voce digitou "BR"- Brasilia até Rio de janeiro')
        break
      elif rt == 'BS':
        rota += 1.2
        print('Voce digitou "BS"- Brasilia até São paulo')
        break
      elif rt == 'RS':
        rota += 1
        print('Voce digitou "RS"- Rio de janeiro até São Paulo')
        break
      elif rt == 'SR':
        rota += 1
        print('Voce digitou "SR"- São paulo ate Rio de Janeiro')
        break
      elif rt == 'SB':
        rota += 1.2
        print('Voce digitou "SB"- São paulo ate Brasilia')
        break
      elif rt == 'RB':
        rota += 1.5
        print('Voce digitou "RB"- Rio de Janeiro ate Brasilia')
        break


rota = 0
rotaObjeto()


total = vd * peso * rota

print(f'Total a pagar(R$):{total}(dimensao:{vd} * peso: {peso} * rota {rota} ')



```
