desempacotamento de parametros em funcoes

```python
#* significa desempacotamento ele basicamente geral uma tupla dentro de num e todo numero colocado nele sera adicionado

  soma = 0
def soma(*num): 
  print(f'Tuplas{num}')
  for i in num:
    soma += i 
  return soma

print(f'Resultado:{soma(1,3)} \n')
```
